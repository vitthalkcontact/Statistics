# Normalization Vs Standardization

### Normalization (min-max Normalization)

In this approch we will scale down the values of the features between 0 and 1

```python
from IPython.display import Image
Image("min_max_scalar.png")
```

```python
import pandas as pd
df = pd.read_csv('wine.csv', header=None, usecols=[0,1,2], skiprows=1)
df.columns = ['Class', 'Alcohol', 'Malic']
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
```

```python
scaling.fit_transform(df[['Alcohol', 'Malic']])
```

# Standardization (Z-score Normalization)

Here all the features will be transformed in such a way that it will have a properties of a standard normal distribution with mean Zero and standard deviation 1

```python
from IPython.display import Image
Image("standard_scalar.Gif")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaling = StandardScaler()
```

```python
scaling.fit_transform(df[['Alcohol', 'Malic']])
```
